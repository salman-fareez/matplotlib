```python
import matplotlib.pyplot as plt
```

line plot

```python
sales = [75000, 78000, 82000, 85000, 87000, 90000, 93000, 95000, 100000, 105000, 110000, 115000]
```

```python
plt.plot(sales);
```

### customize x axis

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
plt.plot(months,sales)
```

```python
plt.plot(months,sales,marker='*')
```

```python
plt.plot(months,sales,marker='*',color='r')
```

### customize labels

```python
plt.plot(months,sales,marker='*',color='r')
plt.xlabel('month')
plt.ylabel('sales')
plt.title('month over sales')
```

#### customize size

```python
plt.figure(figsize=(20,8))
plt.plot(months,sales,marker='*',color='r')
plt.xlabel('month')
plt.ylabel('sales')
plt.title('month over sales');
```

```python
sales2 = [65000, 74000, 79000, 84000, 87000, 90000, 95000, 97000, 100000, 106000, 130000, 145000]
```

```python
plt.figure(figsize=(20,8))
plt.plot(months,sales,marker='*',color='r')
plt.plot(months,sales2,marker='s',color='g')
plt.xlabel('month')
plt.ylabel('sales')
plt.title('month over sales');
```

#### Barplot

```python
Units_Sold= [20, 35, 15, 40, 30, 25, 45]
Units_Sold1= [25, 40, 45, 50, 40, 25, 55]
Product= ['Product A', 'Product B', 'Product D', 'Product E', 'Product F', 'Product G', 'Product H']
```

```python
plt.bar(Product,Units_Sold)
plt.xticks(rotation=90);
```

```python
plt.bar(Product,Units_Sold)
plt.grid()
plt.xticks(rotation=90);
```

```python
plt.barh(Product,Units_Sold,color=['Red'])
plt.xticks(rotation=90);
```
